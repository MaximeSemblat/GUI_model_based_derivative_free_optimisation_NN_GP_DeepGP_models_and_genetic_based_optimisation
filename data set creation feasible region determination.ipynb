```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import qmc

def sobol_sampling(bounds, n_samples):
    """
    Generate samples using Sobol Sampling.

    Parameters:
    - bounds (list of tuples): List of tuples specifying the lower and upper bounds for each dimension.
    - n_samples (int): Number of samples to generate.

    Returns:
    - samples (numpy array): Array containing the generated samples.
    """
    n_dim = len(bounds)
    samples = qmc.Sobol(n_dim).random(n_samples)

    # Scale the samples to the given bounds
    for i in range(n_dim):
        samples[:, i] = samples[:, i] * (bounds[i][1] - bounds[i][0]) + bounds[i][0]

    return samples

# Define the boundaries for x1 and x2
bounds = [(50, 500), (2000, 6000)]
n_samples = 100

# Generate samples using Sobol Sampling
samples = sobol_sampling(bounds, n_samples)



# samples = np.array([
#     [129.62383679, 3191.16923586],
#     [346.60005379, 4955.18548787],
#     [442.29042819, 2100.83406419],
#     [209.71984798, 5865.83305523],
#     [257.54726701, 2593.68543327],
#     [478.72638819, 5326.36607066],
#     [282.8355372, 3745.5435805],
#     [52.65513714, 4479.19555753],
#     [81.50484655, 2428.77472192],
#     [310.34650914, 5662.89239749],
#     [449.98026569, 3269.10563186],
#     [230.15404404, 4502.24909186],
#     [181.30348967, 3792.23045334],
#     [414.347634, 4057.50868469],
#     [374.90059431, 2890.37561417],
#     [157.46411695, 5154.67588976],
#     [135.25839099, 2816.37207419],
#     [368.98233783, 5080.24194464],
#     [392.60713384, 3976.04321316],
#     [174.91657417, 4240.88340998],
#     [239.61656815, 3468.60420331],
#     [468.64365521, 4701.18803531],
#     [320.21933412, 2370.45696378],
#     [99.75450197, 5604.02264073],
#     [71.31736996, 3553.85750905],
#     [292.29739411, 4288.06911409],
#     [496.97838048, 2644.19069141],
#     [267.42279767, 5377.42332742],
#     [203.80076338, 2167.02510417],
#     [420.08349188, 5932.45450035],
#     [340.21579032, 3015.16897604],
#     [107.88551257, 4779.62326258],
#     [114.50732672, 2027.16789022],
#     [333.4816989, 5759.8118335],
#     [427.64085382, 3186.83319539],
#     [196.13455739, 4920.45892403],
#     [274.86788235, 3632.80932605],
#     [489.64063781, 4396.61717787],
#     [298.81036421, 2534.66795012],
#     [64.91517508, 5299.44618791],
#     [92.31606917, 3358.4003523],
#     [327.55042491, 4623.89176711],
#     [462.13733698, 2448.73173907],
#     [246.01211115, 5713.24798465],
#     [168.30184106, 2987.48132214],
#     [399.33414422, 5221.6328159],
#     [361.43205692, 3835.6269598],
#     [142.91751594, 4068.79954413],
#     [151.06154657, 3902.24913508],
#     [381.41404721, 4134.98988375],
#     [407.00951582, 2811.91979721],
#     [188.74905715, 5045.63182592],
#     [222.48821356, 2257.60650262],
#     [457.53714484, 5521.560736],
#     [303.61279324, 3409.45878625],
#     [88.1261779, 4674.39581826],
#     [60.31388659, 2733.60568658],
#     [275.28573908, 5498.94593656],
#     [485.45302306, 3573.9383921],
#     [250.9330167, 4338.30062672],
#     [216.11576648, 3112.39852011],
#     [435.78362716, 4846.4560844],
#     [353.93152009, 2210.54191515],
#     [122.1849212, 5943.62532347],
#     [125.36802813, 2503.414765],
#     [357.78067987, 5268.19923893],
#     [430.61541566, 3664.06245157],
#     [214.25031316, 4427.86418647],
#     [249.93784046, 3155.58582917],
#     [479.41099564, 4889.20639455],
#     [281.76640994, 2058.41519684],
#     [60.86881342, 5791.06442258],
#     [89.11448936, 3866.87588319],
#     [309.6616888, 4100.05075485],
#     [451.0496092, 2956.23245835],
#     [221.94015486, 5190.38154557],
#     [185.55908208, 2479.98241335],
#     [403.16722081, 5744.49720606],
#     [386.57539058, 3327.14973763],
#     [152.93386467, 4592.6424861],
#     [146.98812058, 3378.21194157],
#     [364.39715559, 4643.14279705],
#     [396.80779628, 2288.8532877],
#     [163.7910943, 5552.81381682],
#     [240.63097835, 2780.66707402],
#     [460.48469599, 5014.3844448],
#     [327.8839069, 3933.50179866],
#     [99.01474831, 4166.23732448],
#     [70.30317266, 2241.79308116],
#     [300.45613707, 5974.87402335],
#     [489.31402061, 3081.1474137],
#     [268.16233508, 4815.20732492],
#     [192.07082088, 3605.18774763],
#     [424.66889038, 4369.55137551],
#     [336.01491498, 2702.35639066],
#     [119.01120869, 5467.69512817],
#     [111.73467403, 3714.29087222],
#     [343.39889893, 4447.94813171],
#     [418.21804023, 2624.93808195],
#     [198.63255355, 5357.61355609]
# ])

# Visualize the data coverage
plt.figure(figsize=(8, 6))
plt.scatter(samples[:, 0], samples[:, 1], c='blue', marker='o', label='Data Points')
plt.xlabel('UASB Volume (m^3)')
plt.ylabel('Influent COD (g.m^-3)')
plt.title('Data Coverage (Sobol Sampling)')
plt.grid(True)
plt.legend()
plt.show()
```
